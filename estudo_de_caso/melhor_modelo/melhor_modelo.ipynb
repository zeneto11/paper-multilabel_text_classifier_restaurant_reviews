# Melhor modelo - Logistic Regression com stemming e 100% de labels

## Importar bibliotecas

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.multiclass import OneVsRestClassifier
```

## Carregar datasets

```python
# Carregar dataset
df_fold1 = pd.read_csv('/content/dataset-fold_1-multilabel.csv', encoding='utf-8')
df_fold2 = pd.read_csv('/content/dataset-fold_2-multilabel.csv', encoding='utf-8')
df_fold3 = pd.read_csv('/content/dataset-fold_3-multilabel.csv', encoding='utf-8')
df_fold4 = pd.read_csv('/content/dataset-fold_4-multilabel.csv', encoding='utf-8')
df_fold5 = pd.read_csv('/content/dataset-fold_5-multilabel.csv', encoding='utf-8')
```

## Treinar modelo

```python
# Dividir o dataset
df = pd.concat([df_fold1, df_fold2, df_fold3, df_fold4, df_fold5])
train_data = df.dropna()

X_train = train_data['sentenca']
y_train = train_data.drop('sentenca', axis=1)
```

```python
# Converta as sentenças em vetores de recursos TF-IDF usando o TfidfVectorizer:
vectorizer = TfidfVectorizer(min_df=1, ngram_range=(1, 2))
X_train_tfidf = vectorizer.fit_transform(X_train)
```

```python
# Criar e treinar modelo
model = OneVsRestClassifier(LogisticRegression(C=10.0, max_iter=1000, penalty='l1', solver='liblinear', multi_class='ovr'))
model.fit(X_train_tfidf, y_train)
```

## Exportar modelo

```python
import joblib
joblib.dump(model, 'artigo-melhor_modelo.pkl')

from google.colab import files
files.download('artigo-melhor_modelo.pkl')
```
